### Q. 1) For 'Kota Factory', who were the cast and what is the release date?

### Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
plt.figure(figsize=(26,6))
sns.countplot(df['release_year'],order=df['release_year'].value_counts().index,palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

We could observe that highest number of TV show & Movies were released in the year 2018 followed by 2017 , 2019 , 2020 and before 2000 least number of movies & tv shows were released.

### Q. 3) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
print(df['type'].value_counts())
sns.countplot(df['type'],palette='viridis')
plt.show()
```

Total 6131 Movies were released & 2676 Tv shows were released. Compare to Tv shows maximum movies were released.

### Q. 4) Show only the Titles of all Movies that were released in India.

```python
df[(df['country']=='India') & (df['type']=='Movie')].title.to_frame()
```

### Q. 5) Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?

```python
a = df['director'].value_counts().to_frame().head(10)
plt.figure(figsize=(20,6))
sns.barplot(a.index,a.director,palette='viridis')
plt.show()
```

### Q. 6) Which are the top 10 country that has the Highest No. of TV Shows ?

```python
a = df[df['type']=='TV Show'].country.value_counts().to_frame().head(10)
plt.figure(figsize=(16,6))
sns.barplot(a.index,a.country,palette='viridis')
plt.show()
```

### Q. 7) Which are the top 10 country that has the Highest No. of Movies ?

```python
a = df[df['type']=='Movie'].country.value_counts().to_frame().head(10)
plt.figure(figsize=(16,6))
sns.barplot(a.index,a.country,palette='viridis')
plt.show()
```
